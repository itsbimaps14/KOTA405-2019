```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Hyperparameter space
batch_sizes = ['16', '32']
learning_rates = [2, 3, 5]
epochs = [5, 25, 50]

# Create meshgrid
X, Y, Z = np.meshgrid(batch_sizes, learning_rates, epochs)

# Calculate grid search results (dummy data)
accuracy_scores = np.random.rand(len(batch_sizes), len(learning_rates), len(epochs))

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot grid search results
ax.scatter(X.flatten(), Y.flatten(), Z.flatten(), c=accuracy_scores.flatten(), cmap='viridis')

# Set labels and title
ax.set_xlabel('Batch Size')
ax.set_ylabel('Learning Rate ("y"e-5)')
ax.set_zlabel('Epochs')
ax.set_title('Grid Search Results')

# Set tick labels
#ax.plot(x.astype('str'), y)
ax.set_xticks(batch_sizes)
ax.set_yticks(learning_rates)
ax.set_zticks(epochs)

# Rotate the plot for better view
ax.view_init(elev=20, azim=120)

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

## recreate a DataFrame that looks how a gridsearch would look
epoch = [5, 25, 50]
lr = ['2e-5', '3e-5', '5e-5']
bs = [16, 32]
grid = []
for x in epoch:
    for y in lr:
        for z in bs:
            grid.append([x,y,z])
            
results = pd.DataFrame(data=np.array(grid), columns=["Epoch","Learning Rate","Batch Size"])

## let's assume your mean_test_score gradually increases with some noise 
np.random.seed(42)
mean_scores = 1
results["mean_scores"] = mean_scores

fig = px.scatter_3d(results, x='Epoch', y='Learning Rate', z='Batch Size')

fig.update_layout(
    title="Grid Search Visualization",
    autosize=True, width=1000, height=1000,
    margin = dict(l=65, r=50, b=300, t=90))
fig.show()
```
