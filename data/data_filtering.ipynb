```python
import json
import os

dir_source = 'data_mentah'
dir_destination = 'data_mentah_filtered'

sources = ['Cnbcindonesia-market', 'Bisnis', 'Indopremier - Ipotnews', 'Investor', 'Kontan']

if not os.path.exists(dir_destination):
    os.makedirs(dir_destination)

for file in os.listdir(dir_source):
    if file.endswith('.json') and file != 'ALL.json':
        tmp = []
        print(file)
        f = open(os.path.join(dir_source, file))
        json_file = json.load(f)
        
        for idx in (json_file['data']):
            source_name = idx['meta']['source']['name']
            if source_name in sources:
                tmp.append(idx)
                
        with open(os.path.join(dir_destination, file), 'w') as outfile:
            json.dump(tmp, outfile, indent=1)
        
        f.close()
```

```python
import json
import os
import pandas as pd

dir_destination = 'data_mentah_filtered'
data_list = []

for file in os.listdir(dir_destination):
    if file.endswith('.json') and file != 'ALL.json':
        f = open(os.path.join(dir_destination, file))
        json_file = json.load(f)
        data_list.append(json_file)

        f.close()
        
merged_data = []
for data in data_list:
    merged_data.extend(data)

with open(os.path.join(dir_destination, 'ALL.json'), 'w') as f:
    json.dump(merged_data, f, indent=1)
    print(os.path.join(dir_destination, 'ALL.json')+' created.')

f.close()
```

```python
import json
import pandas as pd

f = open('data_mentah_filtered/ALL.json')
file = json.load(f)

sources = ['Cnbcindonesia-market', 'Bisnis', 'Indopremier - Ipotnews', 'Investor', 'Kontan']
total = [0, 0, 0, 0, 0]
lst = []

for i in file:
    for j in range(len(sources)):
        if (i['meta']['source']['name'] == sources[j]):
            total[j] += 1

sources[0] = 'CNBC Indonesia'
sources[2] = 'Indopremier'
sources[3] = 'Investor Daily'

for i in range(len(sources)):
    lst.append((sources[i], total[i]))

pd.DataFrame(lst, columns=('Sumber', 'Jumlah Berita')).to_csv('jumlah_berita_sumber.csv')
pd.DataFrame(lst, columns=('Sumber', 'Jumlah Berita'))
```
